## Binary Search

```python
def binarySearch(arr, low, high, value):
    if low <= high:
        mid = low + ((high-low)//2)
        if arr[mid] == value:
            return mid
        elif arr[mid] > value:
            high = mid - 1
            return binarySearch(arr, low, high, value)
        else:
            low = mid + 1
            return binarySearch(arr, low, high, value)
    else:
        return -1
```

```python
# Recursion
def binarySearch(arr, low, high, x):
    if high >= low:
        mid = low + (high - low) // 2

        if arr[mid] == x:
            return mid
        elif arr[mid] > x:
            return binarySearch(arr, low, mid-1, x)
        else:
            return binarySearch(arr, mid + 1, high, x)

    else:
        return -1
```

```python
binarySearch([1,2,3,4,5,6,7], 0, 6, 6)
```

```python
def binarySearch(arr, low, high, value, isAsc):
    if low == 0 and high == len(arr)-1:
        if arr[low] < arr[high]:
            isAsc = True
        else:
            isAsc = False
    if low <= high:
        mid = low + ((high-low)//2)
        if arr[mid] == value:
            return mid
        if isAsc:
             if arr[mid] > value:
                high = mid - 1
                return binarySearch(arr, low, high, value, isAsc)
             else:
                low = mid + 1
                return binarySearch(arr, low, high, value, isAsc)
        else:
            if arr[mid] < value:
                high = mid - 1
                return binarySearch(arr, low, high, value, isAsc)
            else:
                low = mid + 1
                return binarySearch(arr, low, high, value, isAsc)
    return -1
```

```python
def binarySearch(arr, low, high, value):
    while low <= high:
        mid = (low + high)//2
        if arr[mid] == value:
            return mid

        if arr[mid] <= value:
            low = mid+1
        else:
            high = mid - 1
    return -1
```

### Find the first occurence of an element in an array

```python
result = -1
def binarySearchToFindFirstOccIndex(arr, start, end, value):
    global result
    if start <= end:
        mid = (start + end)//2
        if arr[mid] == value:
            result = mid
            end = mid-1
            return binarySearch(arr, start, end, value)
        elif arr[mid] < value:
            start = mid+1
            return binarySearch(arr, start, end, value)
        else:
            end = mid - 1
            return binarySearch(arr, start, end, value)
    return result
```

```python
binarySearchToFindFirstOccIndex([-2, -1, 0, 0, 3, 4, 5], 0, 6, 0)
```

## Find the Last occurence of an element in an array

```python
result = -1
def binarySearchToFindLastOccIndex(arr, start, end, value):
    global result
    if start <= end:
        mid = (start + end)//2
        if arr[mid] == value:
            result = mid
            start = mid+1
            return binarySearch(arr, start, end, value)
        elif arr[mid] < value:
            start = mid+1
            return binarySearch(arr, start, end, value)
        else:
            end = mid - 1
            return binarySearch(arr, start, end, value)
    return result

binarySearchToFindFirstOccIndex([-2, -1, 0, 0, 3, 4, 5], 0, 6, 0)
```

### Find first non duplicating element in an array

```python
def binarySearchToFindFirstNonDuplicateEle(arr, start, end):
    while start < end:
        mid = (start + end) // 2

        if mid % 2 != 0:
            mid -= 1

        if arr[mid] == arr[mid + 1]:
            start = mid + 2 
        else:
            end = mid

    return start
```

## Merge Sort

```python
def merge(arr1, arr2):
  i = j = 0
  result = []
  while i < len(arr1) and j < len(arr2):
    if arr1[i] <= arr2[j]:
      result.append(arr1[i])
      i += 1
    else:
      result.append(arr2[j])
      j += 1

  return result + arr1[i:] + arr2[j:]

def mergeSort(arr):
  if len(arr) <= 1:
    return arr

  mid = len(arr) // 2
  leftArr = arr[:mid]
  rightArr = arr[mid:]

  leftSortedArr = mergeSort(leftArr)
  rightSortArr = mergeSort(rightArr)

  return merge(leftSortedArr, rightSortArr)


arr1 = [2,4,6]
arr = [1,3,7,8,9,5,3]
# print(mergeSort(arr))
```

```python
arr1 = [2,4,6]
arr = [1,3,7,8,9,5,3]
# print(mergeSort(arr))

def quickSort(arr):
  if len(arr) <=1:
    return arr
  pivot = arr.pop()
  items_lower = []
  items_greater = []

  for i in arr:
    if i < pivot:
      items_lower.append(i)
    else:
      items_greater.append(i)
  return quickSort(items_lower) + [pivot] + quickSort(items_greater)

print(quickSort(arr))
```
