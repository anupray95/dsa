### Minimum Coins required to make a sum

```python
def func(ip, req, coinCount = 0, start=0):
    if req == 0:
        return 0
    if req < 0:
        return float('inf')
    min_coins = float('inf')
    for i in range(start, len(ip)):
        if(ip[i] <= req):
            count = 1 + func(ip, req-ip[i],coinCount+1, i)
            min_coins = min(count, min_coins)
    return min_coins
print(
func([1,2,5], 5))
```

### Knapsack maximum profit for a given weight

```python
def knapsack(weights, values, W, n):
    if  n == 0 or W == 0:
        return 0
    if weights[n-1] <= W:
        return max(values[n-1] + knapsack(weights, values, W-weights[n-1], n-1),
        knapsack(weights, values, W, n-1))
    return knapsack(weights, values, W, n-1)


weights = [2, 3, 4, 5]
values  = [3, 4, 5, 6]
W = 5
n = len(weights)
print(
knapsack(weights, values, W, n)
)
```

### Backtracking - All permuntation of a string

```python
def permute(string, start=0, arr=[]):
    if(start == len(string)-1):
        arr.append(string)

    s = set()
    for i in range(start, len(string)):
        if string[i] not in s: # do not repeat for same 
            s.add(string[i])
            string[start], string[i] = string[i],string[start]
            permute(string, start+1, arr)
            string[start], string[i] = string[i],string[start] #backtracking
    return arr
```

```python
permute(list('aac'))
```

```python
weights = [1,2,3,4,5,6]
values  = [1,2,3,4,5,6]
W = 10
n=len(weights)

#create a matrix and counter(counter is not necessary, just to track function call count)
matrix = [[-1]*(W+1) for _ in range(n+1)]
counter = 0
def knapsack(weights, values, W, n):
    global counter
    global matrix
    
    if  n == 0 or W == 0:
        return 0

    # return if already calculated
    if(matrix[n][W] != -1):
        return matrix[W][n]
    counter += 1
    
    if weights[n-1] <= W:
        # store in matrix first, then return
        matrix[n][W] = max(values[n-1] + knapsack(weights, values, W-weights[n-1], n-1),
        knapsack(weights, values, W, n-1))
        return matrix[n][W]
    matrix[n][W] = knapsack(weights, values, W, n-1)
    return matrix[n][W]



print(knapsack(weights, values, W, len(weights)))
print('Total recursive calls: ', counter)
```

```python
from pprint import pprint
weights = [2, 3, 4, 5]
values  = [3, 4, 5, 6]
W = 5
n = len(weights)

matrix = [[None]*(W+1) for _ in range(n+1)]

def topDownKnapSack(weights, values, W, n):
    for i in range(n+1):
        for j in range(W+1):
            if i == 0 or j == 0:
                matrix[i][j] = 0

    for i in range(1, n+1):
        for j in range(1, W+1):
            if weights[i-1] <= j:
                matrix[i][j] = max(
                    values[i-1] + matrix[i-1][j-weights[i-1]],
                    matrix[i-1][j]
                )
            else:
                matrix[i][j] = matrix[i-1][j]

    pprint(matrix)
    return matrix[n][W]
topDownKnapSack(weights, values, W, n)
```

## Subset Sum problem (return True, if subset sum exists)

```python
def subsetSum(arr, summ, n):
    if summ == 0:
        return True
    if n == 0:
        return False
    if arr[n-1] <= summ:
        include =  subsetSum(arr, summ-arr[n-1], n-1) 
        exclude = subsetSum(arr, summ, n-1)
        return include or exclude
    return subsetSum(arr, summ, n-1)
        
```

```python
arr = [2, 3, 7, 9, 10]
summ = 11

subsetSum(arr, summ, len(arr))
```

#### subsetSum with memoization

```python
arr = [2, 3, 7, 9, 10]
summ = 11
n = len(arr)

matrix = [[-1]*(summ+1) for _ in range(n+1)]
def subsetSum(arr, summ, n):
    global matrix
    if summ == 0:
        return True
    if n == 0:
        return False
    if matrix[n][summ] != -1:
        return matrix[n][summ]
    if arr[n-1] <= summ:
        include =  subsetSum(arr, summ-arr[n-1], n-1) 
        exclude = subsetSum(arr, summ, n-1)
        matrix[n][summ] = include or exclude
        return matrix[n][summ]
    matrix[n][summ] = subsetSum(arr, summ, n-1)
    return matrix[n][summ]
        
```

```python
arr = [2, 3, 7, 9, 10]
summ = 11
n = len(arr)
subsetSum(arr, summ, len(arr))
```

### recursion: Target sum(can add or substract a number)

```python
def func(arr, op, n, summ):
    if n == 0:
        if op == summ:
            return 1
        return 0
    return func(arr, op+arr[n-1], n-1, summ) + func(arr, op-arr[n-1], n-1, summ)
```

```python
arr = [1, 2, 5]
summ = 5
n = len(arr)
def func(arr, n, summ):
    if summ == 0:
        return 1
    if n == 0:
        return 0

    if arr[n-1] <= summ:
        return  func(arr, n, summ - arr[n-1]) + func(arr, n-1, summ)
    return func(arr, n-1, summ)
```

```python
matrix = [[None]*(summ+1) for _ in range(n+1)]
def func(arr, n, summ):
    global matrix
    for i in range(n+1):
        for j in range(summ + 1):
            if j == 0:
                matrix[i][j] = 1
            if i == 0:
                matrix[i][j] = 0
            if i == 0 and j == 0:
                matrix[i][j] = 1
    for i in range(1, n+1):
        for j in range(1, summ+1):
            if arr[i-1] <= j:
                matrix[i][j] = matrix[i][j - arr[i-1]] + matrix[i-1][j]
            else:
                matrix[i][j] = matrix[i-1][j]
            
    return matrix[n][summ]
                
```

### rod cutting problem

#### recursion

```python
def func(length, price, summ, n):
    if n == 0 or summ == 0:
        return 0
    if length[n-1] <= summ:
        return max(price[n-1] + func(length, price, summ - length[n-1], n), func(length, price, summ, n-1))
    else:
        return func(length, price, summ, n-1)
```

```python
length = [1, 2, 3, 4, 5, 6, 7, 8]
price  = [1, 5, 8, 9, 10, 17, 17, 20]
summ = 8
n = len(length)
func(length, price, summ, n)
```

#### dynamic programming

```python
matrix = [[None]*(summ+1) for _ in range(n+1)]
def func(length, price, summ, n):
    for i in range(n+1):
        for j in range(summ+1):
            if i == 0 or j == 0:
                matrix[i][j] = 0
    for i in range(1,n+1):
        for j in range(1,summ+1):
            if length[i-1] <= j:
                matrix[i][j] =  max(price[i-1] + matrix[i][j - length[i-1]], matrix[i-1][j])
            else:
                matrix[i][j] =   matrix[i-1][j]
    return matrix[n][summ]
```

### Minimum no of coins to make a given sum

#### recursion

```python
def func(arr, summ, n):
    if summ == 0:
        return 0
    if n == 0:
        return float('inf')
    if arr[n-1] <= summ:
        return min(1+ func(arr, summ - arr[n-1], n), func(arr, summ, n-1))
    else:
        return func(arr, summ, n-1)
        
```

```python
arr = [1, 2, 5]
summ = 11
n = len(arr)
func(arr, summ, n)
```
