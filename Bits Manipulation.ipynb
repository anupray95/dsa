## Set kth value to 1 (0 based index)

```python
num = 0b1001

k = 2
result = num | 1 << k

bin(result)
```

## Get kth value 

```python
def getKthValue(num, k):
    temp = 1 << k 
    
    if num & temp > 0: # if 1 is at kth place, & operation will return >0 value, else 0
        return 1
    else:
        return 0

num = 0b1101

print(getKthValue(num, 2))
print(getKthValue(num, 1))
```

## Clear kth Value

```python
'''
xor table
=========
    
1 1 0  # xor with 1, return 0
0 1 1
0 0 0  # xor with 0, return same num
1 0 1

perfect operator to clear a bit
'''

pass
```

```python
def clearKthValue(num, k):
    temp = 1 << k
    print(bin(num ^ temp))

num = 0b1101
clearKthValue(num, k)
```

## Toggle Kth Value

```python
'''
xor table
=========
    
1 1 0  # xor with 1, toggles value to 0
0 1 1  # xor with 1, toggles value to 1
0 0 0  
1 0 1

perfect operator to toggle a bit

Logic
=====
1. create 111.. as same bit as the original number
2. do xor with original number
'''

pass
```

```python
def toggleBit(num, k):
    temp = 1 << k
    return num ^ temp
    
num  = 0b1101
k = 2

result = toggleBit(num, k)
print(bin(result))
```
